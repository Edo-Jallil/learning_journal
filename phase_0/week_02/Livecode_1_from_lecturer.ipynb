## **Quest 2**

- Buatlah fungsi untuk menghitung jumlah kasus harian Covid-19!
- (Hint: Kamu bisa menggunakan .diff() untuk menghitung perbedaan tiap baris data/menghitung jumlah kasus harian 
- dan .transpose() untuk mengubah kolom menjadi baris atau sebaliknya)

```python
df.T.head()
```

```python
df_t = df.T
df_t.rename(columns=df_t.iloc[1], inplace=True)
df_t.drop(['Province/State', 'Country/Region', 'Lat', 'Long'], axis=0, inplace=True)
df_t.head()
```

```python
df.head()
```

```python
def kasus_harian(df, kolom):
  df_t = df.T
  df_t.rename(columns=df_t.iloc[1], inplace=True)
  df_t.drop(['Province/State', 'Country/Region', 'Lat', 'Long'], axis=0, inplace=True)

  harian = df_t[kolom].diff()
  index = df_t.index

  df_harian = pd.DataFrame({'date': index,
                     kolom: harian}).reset_index(drop=True)
  return df_harian
```

```python
kasus_harian(df, 'Indonesia')
```

## **Quest 4**

- Gunakan for loop dan conditional if untuk membuat sebuah list yang berisikan jumlah kasus harian Covid-19
- dengan ketentuan hanya mengambil data Indonesia saja
- dan jumlah kasus di bawah 10000.
- Berapa jumlah hari dengan kasus di bawah 10000?

```python
ind
```

```python
# loop for
case_under10k = []

for i in ind['Indonesia']:
  if i < 10000:
    case_under10k.append(i)

len(case_under10k)
```

```python
# loop while
counter = 0
kasus_u10k = []

while counter < len(ind['Indonesia']):
  if ind.iloc[counter, 1] < 10000:
    kasus_u10k.append(ind.iloc[counter, 1])

  counter += 1

len(kasus_u10k)
```
